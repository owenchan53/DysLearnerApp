```python
import time
```

```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization")
```

```python
def menu():
    print("\n Welcome to AIggregator!\n")
    print("                  _______")
    print("         _ _ _ ( o)> /")
    print("        /_ __ (_____/ ")
    print("            /      \\")
    print("           /   --   \\")
    print("          /  _    _ \\\ ")
    print("         /  (_)  (_) \\")
    print("        |             |")
    print("        \\_\\_________/_/")
    print("         (___________)\n\n")
    print("Press Enter to Proceed:\n\n", end="")
    input()

    choice = -1  # Initialize choice to a non-zero value

    while (choice != 0):
        print("\nPlease choose one of the following options:")
        print("1) What is Aggregator")
        print("2) How does it work?")
        print("3) Summarize")
        print("0) Quit\n")

        choice = int(input())

        if choice == 1:
            print("\nAggregator is a software or tool that collects and combines data from multiple sources into a single location.\n")
        elif choice == 2:
            print("\nAggregator works by using APIs (Application Programming Interfaces) to collect data from different sources, then organizing and presenting the data in a meaningful way.\n")
        elif choice == 3:
            ARTICLE = input("Enter the article you want to summarize: ")


            print("Generating summary...\n")

            summary = summarizer(ARTICLE, max_length=250, min_length=25, do_sample=False)
            print("\nHere is the summary of the article:\n", summary[0]['summary_text'], "\n")

        elif choice == 0:
            print("Exiting Program...")
            print("Program Terminated.\n")
        else:
            print("Invalid choice. Please choose option 1, 2, or 3.\n")

menu()
```
